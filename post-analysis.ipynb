```python
from abtem.measure import Measurement
```

```python
import numpy as np
import glob
import os
```

```python
for file in glob.glob('*.hdf5'):
    name,_ = file.split('.')
    measurement = Measurement.read(file)
    measurement.save_as_image(name+'.png')
```

```python
%matplotlib notebook
```

```python
hdf5file = 'raw_data/POSCAR_1000K_1_Zaxis110.hdf5'
measurement = Measurement.read(hdf5file)
```

```python
measurement.show(cmap="inferno")
```

```python
path, fn = os.path.split(hdf5file)
bn, ext = os.path.splitext(fn)
image_path = os.path.join(path, bn+".png")
```

```python
measurement.save_as_image(image_path)
```
