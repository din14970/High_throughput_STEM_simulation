# Mass data export 10 nm cells

```python
import glob
```

```python
potentials_fp = glob.glob("raw_data/*_10nm_*_potential.hdf5")
sims_fp = glob.glob("raw_data/*_10nm_*0.hdf5")
```

```python
potentials_fp
```

```python
sims_fp
```

```python
from abtem.measure import Measurement
import os

for file in sims_fp:
    name,_ = os.path.splitext(file)
    measurement = Measurement.read(file)
    print(f"Exporting {name}.png")
    measurement.save_as_image(name+'.png')
```

```python
from abtem.potentials import PotentialArray

for file in potentials_fp:
    pa = PotentialArray.read(file)
    name,_ = os.path.splitext(file)
    print(f"Exporting {name}.png")
    pa.project().save_as_image(name+".png")
```
